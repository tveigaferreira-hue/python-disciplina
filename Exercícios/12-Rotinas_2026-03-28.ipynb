Desta forma, quando chamada, a função retorna tanto o array, como os valores maiores que 10 e a quantidade desses valores.

## 04 – Escreva o programa em Python:

```python
import numpy as np
```

```python
def analisar_valores():
    valores = np.random.normal(loc=10, scale=3, size=5)

    maiores_que_10 = valores[valores > 10]
    
    quantidade = len(maiores_que_10)
    
    return valores, quantidade, maiores_que_10

valores, qtd, maiores = analisar_valores()

print("Valores gerados:", valores)
print("Quantidade > 10:", qtd)
print("Valores > 10:", maiores)
```

## 05 – Seria possível, com o racional proposto, saber quantos e quais valores são superiores a 20? Porque?

Sim, porque a mesma forma utilizada na filtragem pode ser aplicada com uma nova condição (valores > 20).

```python
def analisar_valores():
    maiores_que_20 = valores[valores > 20]
    
    quantidade = len(maiores_que_20)
    
    return valores, quantidade, maiores_que_20

valores, qtd, maiores = analisar_valores()

print("Valores gerados:", valores)
print("Quantidade > 20:", qtd)
print("Valores > 20:", maiores)
```
