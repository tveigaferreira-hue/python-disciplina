## 02 – Descreva o racional da sua resposta:

R.: Foi primeiramente criada uma função que gera um array com distribuição normal (média 10, desvio padrão 3) contendo 5 valores.Em sequencia,  utilizei um filtro booleano para selecionar os valores superiores a 10, e o len() para retornar quantos valores passaram. Desta forma, quando chamada, a função retorna tanto o array, como os valores maiores que 10 e a quantidade desses valores. 

## 03 – Desenhe o fluxograma da resposta:

## 04 – Escreva o programa em Python:

```python
import numpy as np
```

```python
def analisar_valores():
    valores = np.random.normal(loc=10, scale=3, size=5)

    maiores_que_10 = valores[valores > 10]
    
    quantidade = len(maiores_que_10)
    
    return valores, quantidade, maiores_que_10

valores, qtd, maiores = analisar_valores()

print("Valores gerados:", valores)
print("Quantidade > 10:", qtd)
print("Valores > 10:", maiores)
```

## 09 – Desenvolva um programa de computador que retorne a divisão de dois objetos simples. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
# Exemplo
a = 20 
b = 4
```

```python
def divisao(a, b):
    if type(a) not in (int, float) or type(b) not in (int, float):
        return f"Tipos inválidos: {type(a).__name__} e {type(b).__name__}"
    
    if b == 0:
        return "Não é possível dividir por zero"
    
    return a / b
```

```python
print(divisao(a, b))
```

## 10 – Desenvolva um programa de computador que retorne a divisão de dois objetos compostos. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
# Exemplo 
a = [10, 5, 4]
b = [2, 5, 1]
```

```python
def divisao_composto(a, b):
    if type(a) not in (list, tuple) or type(b) not in (list, tuple):
        return "Ambos devem ser listas ou tuplas"
    
    if len(a) != len(b):
        return "As estruturas devem ter o mesmo tamanho"
    
    resultado = []
    
    for x, y in zip(a, b):
        if type(x) not in (int, float) or type(y) not in (int, float):
            return f"Tipos inválidos: {type(x).__name__} e {type(y).__name__}"
        
        if y == 0:
            return "Não é possível dividir por zero"
        
        resultado.append(x / y)
    
    return resultado
```

```python
print(divisao_composto(a, b))
```
