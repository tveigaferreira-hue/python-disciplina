## Instancie um objeto da classe lista contendo 5.000 valores inteiros iniciando em 0. 
Quantos são pares e quantos são impares?

## 01 – Descreva o racional da sua resposta:

R.: Inicio o desafio instanciando um objeto da classe lista. Esta lista precisa ter 5.000 valores e iniciar em 0, desta forma utilizo range(5000) para garantir esse conteúdo na lista. Para responder o número de pares e impares, crio uma estrutura de repetição onde: para cada numero da lista, respondo algo. Desta maneira, passando por todos os números e dividindo em duas categorias: par e impar e somando ao valor da categoria. Para isso utilizo o resultado da sobra da divisão por 2, sendo igual zero, o valor é par. 

## 02 – Desenhe o fluxograma da resposta:

## 03 – Escreva o programa em Python:

```python
lista_questao1 = list(range(5000))
```

```python
pares = sum(1 for n in lista_questao1 if n % 2 == 0)
impares = sum(1 for n in lista_questao1 if n % 2 != 0)

print("Pares:", pares)
print("Ímpares:", impares)
```

## 04 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

Desta vez utilizo o sum() somando os valores que se enquadrarem aos critérios de pares (resto da divisão=0) ou na categoria de impar (resto ds divisão diferente de 0).

```python
pares = sum(1 for n in lista_questao1 if n % 2 == 0)
impares = sum(1 for n in lista_questao1 if n % 2 != 0)

print("Pares:", pares)
print("Ímpares:", impares)
```

## Importe o .csv exportado na última aula e instancie um objeto da classe DataFrame.

Dos números de letras nos genes, quais são pares e quantos são impares?

```python
import pandas as pd
genes0607 = pd.read_csv('genes0607.csv')
genes0607 = genes0607.rename(columns={'2 primeiros caracteres': 'prefixo'})
genes0607
```

## 05 – Descreva o racional da sua resposta:

R.: O racional é semelhante a questão anterior, instanciarei dois objetos: pares (onde da coluna intitulada 'numero de caracteres' ele vai identificar se o resto da divisão de cada valor por 2 é 0, caso seja, soma ao número total), e impares (mesma lógica dos pares, porém com resto da divisão por 2 diferente de 0). A contagem desses valores retorna como booleano, onde true=1, então somando 1 cada vez que o número se enquadrar na equação. Enquanto ainda restar números para "categorizar", o codigo continua respondendo as perguntas. 

## 06 – Desenhe o fluxograma da resposta:

## 07 – Escreva o programa em Python:

```python
pares = (genes0607['numero de caracteres'] % 2 == 0).sum()
impares = (genes0607['numero de caracteres'] % 2 != 0).sum()

print("Pares:", pares)
print("Ímpares:", impares)
```

## 08 – Substitua a estrutura de repetição e responda novamente a pergunta anterior.

Para essa substituição utilizei o iloc() para selecionar a primeira coluna, verifiquei igualmente a questão anterior o resto da divisão dos valores por 2 e criei booleanos, transformando a coluna em categorias (0=pares e 1=impares). O .value_counts() faz a conta dos resultados. 

```python
contagem = (genes0607.iloc[:, 1] % 2).value_counts()

pares = contagem[0]
impares = contagem[1]

print("Pares:", pares)
print("Ímpares:", impares)
```

## 10 – Selecione qualquer uma das instâncias contidas na DataFrame. É possível recuperar as chaves e os valores com ao menos 3 estruturas de repetição diferentes?

Sim, depende da forma que vc espera o resultado.
