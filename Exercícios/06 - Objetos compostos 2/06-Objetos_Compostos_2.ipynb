## 10 – Importe o data.csv (dados de seguro na Flórida em 2012) para uma DataFrame. Armazene em um dicionário o número de entradas, uma lista com as classes das colunas e outra lista com o número de elementos de cada classe, três listas, contendo a média, o desvio padrão e o terceiro quartil da primeira, quarta e nona colunas . Recupere o dicionário.

```python
import os
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
df
```

```python
numero_entradas = df.shape
numero_entradas
```

```python
classe_colunas = df.dtypes
classe_colunas
```

```python
elementos_classe = list(df.dtypes.value_counts())
elementos_classe
```

```python
colunas139 = df.iloc [: , [0,3,8]]
colunas139
```

```python
media = colunas139.mean()
media
```

```python
desvio_padrao = colunas139.std()
desvio_padrao
```

```python
quartil = colunas139.quantile(0.75)
quartil
```

```python
dicionario10 = {
    "numero de entradas": numero_entradas,
    "classe das colunas": classe_colunas,
    "numero de elementos": elementos_classe,
    "media": media,
    "desvio padrao": desvio_padrao,
    "terceiro quartil": quartil
}
dicionario10
```
