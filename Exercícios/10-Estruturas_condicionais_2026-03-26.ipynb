## 04 – Escreva o programa em Python:

```python
import numpy as np
array_random = np.random.normal(loc=10, scale=3, size=5)
valores = []
quantidade = 0

for x in array_random:
    if x > 10:
        valores.append(x)
        quantidade += 1

print("Valores > 10:", valores)
print("Quantidade:", quantidade)
```

Feito essa classificação, utilizo a condicional para responder de acordo com cada tamanho: se nome tiver entre 2 e 4 caracteres devolvo as informações do gene, se tiver <2 caracteres, devolvo os valores nas instâncias onde os números de caracteres dos nomes são superiores a 4, e se tiver >4 caracteres, devolvo o nome dos genes que se incluem nessa categoria.

```python
#Caso o nome tenha entre 2 e 4 caracteres inclusive, posso saber as informações do gene?
caso1 = genes0607[(genes0607['numero de caracteres'] >= 2) & (genes0607['numero de caracteres'] <= 4)]
print("CASO 1:\n", caso1)
```

```python
#Caso tenha menos de 2 caracteres, posso saber todos os valores nas instâncias onde os números de caracteres dos nomes são superiores a 4?
if any(genes0607['numero de caracteres'] < 2):
    caso2 = genes0607[genes0607['numero de caracteres'] > 4]
    print("\nCASO 2:\n", caso2)
```

```python
#caso tenha mais de 4 caracteres, apenas me informe quem é superior
caso3 = genes0607[genes0607['numero de caracteres'] > 4]
print("\nCASO 3:\n", caso3.index.tolist())
```
