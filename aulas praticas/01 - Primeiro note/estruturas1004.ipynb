estruturas de repetição

```python
array1 = np.array([10,20,30,40,50])
array1
```

```python
contador = 0 
while contador < 10: 
    contador += 1
    print(contador)
    
```

```python
contador = 0 
while contador < len(array1):
    print(array1[contador])
    contador += 1
```

```python
for i in range(len(array1)):
    print(array1[i])
```

```python
for i in enumerate(array1):
    print(i)
```

```python
for key, value in enumerate(array1):
    print(key, value)
```
