```python
# soma e subtração
num_1 = 4
num_2 = 2
print(num_1+num_2)
print(num_1-num_2)
```

```python
# objetos compostos
tupla1 = (1, 2, 3, 4)
tupla2 = (1, 2, 3, 4)

print(tupla1+tupla2)
```

```python
#soma uma tupla com ela mesma
tupla1 += tupla2
display(tupla1)
display(tupla2)
```

```python
# transforma tupla em lista
listatupla = list(tupla1)
listatupla
```

```python
#cada vez que da play, faz uma soma (em tupla não da, só na lista)
listatupla[2] += 2
listatupla
```
