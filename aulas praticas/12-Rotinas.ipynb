```python
def contar_inteiros(obj: list | tuple) -> int:
    contador = 0
    for x in range (len(obj)):
        if type(obj[x]) == int:
            contador += 1
    print(f'Quantidade de inteiros: {contador}')
    return contador
```

```python
obj = [1,2,3,4,5,'a','b','c','d']
resultado_contador = contar_inteiros(obj=obj)
print(resultado_contador)
```
